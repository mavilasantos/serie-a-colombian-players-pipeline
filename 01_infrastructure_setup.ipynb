# Infrastructure Setup — Unity Catalog Provisioning
## Colombian Players in Serie A | Medallion Architecture
---

**When to run:** Execute once before running any pipeline for the 
first time. Safe to rerun — all commands use `IF NOT EXISTS`.

**Do not include in the automated Workflow job** — this is a 
one-time setup step, not a recurring pipeline task.

**What to run next:**
1. `02_bronze_ingestion.ipynb` — extracts API data into Bronze
2. LakeFlow Pipeline — transforms Bronze into Silver and Gold

### What Gets Provisioned

| Layer | Catalog | Schema | Asset |
|---|---|---|---|
| Bronze | `bronze_dev` | `api_sports` | `landing_zone` (Volume) |
| Silver | `silver_dev` | `football_analytics` | — |
| Gold | `gold_dev` | `football_analytics` | — |

> Tables are created automatically by the pipeline — this notebook 
> only provisions the catalogs, schemas, and volumes the pipeline 
> depends on.

```python
print("Provisioning Medallion Architecture infrastructure...")

# Bronze: schema named after source system (api_sports)
spark.sql("CREATE CATALOG IF NOT EXISTS bronze_dev")
spark.sql("CREATE SCHEMA IF NOT EXISTS bronze_dev.api_sports")
spark.sql("CREATE VOLUME IF NOT EXISTS bronze_dev.api_sports.landing_zone")
print("Bronze infrastructure ready: bronze_dev.api_sports.landing_zone")

# Silver/Gold: schemas named after business domain (football_analytics)
spark.sql("CREATE CATALOG IF NOT EXISTS silver_dev")
spark.sql("CREATE SCHEMA IF NOT EXISTS silver_dev.football_analytics")
print("Silver infrastructure ready: silver_dev.football_analytics")

spark.sql("CREATE CATALOG IF NOT EXISTS gold_dev")
spark.sql("CREATE SCHEMA IF NOT EXISTS gold_dev.football_analytics")
print("Gold infrastructure ready: gold_dev.football_analytics")

print("\nAll infrastructure provisioned successfully.")
print("Next step: Run 02_bronze_ingestion.ipynb")
```
